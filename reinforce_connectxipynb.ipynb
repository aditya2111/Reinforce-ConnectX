```python
def get_legal_acions(obs, config):
  a = []
  for i in range(config.columns):
    if(obs.board[i] == 0):
      a.append(i)
  return a
```

```python

def is_winning(obs, config):
    """
    Check if there's a winning condition in Connect X game.

    Args:
        obs: Observation object with board attribute (1D list)
        config: Configuration dict with rows, columns, and inarow

    Returns:
        int: Player number (1 or 2) if there's a winner, 0 if no winner
    """
    board = obs.board
    rows = config['rows']
    cols = config['columns']
    inarow = config['inarow']

    def get_cell(row, col):
        """Get cell value from 1D board using 2D coordinates"""
        if 0 <= row < rows and 0 <= col < cols:
            return board[row * cols + col]
        return 0

    def check_direction(row, col, delta_row, delta_col):
        """Check for winning sequence in a specific direction"""
        player = get_cell(row, col)
        if player == 0:
            return False

        count = 1

        # Check forward direction
        r, c = row + delta_row, col + delta_col
        while get_cell(r, c) == player:
            count += 1
            r += delta_row
            c += delta_col

        # Check backward direction
        r, c = row - delta_row, col - delta_col
        while get_cell(r, c) == player:
            count += 1
            r -= delta_row
            c -= delta_col

        return count >= inarow

    # Check all positions on the board
    for row in range(rows):
        for col in range(cols):
            if get_cell(row, col) != 0:
                # Check all four directions from this position
                directions = [
                    (0, 1),   # Horizontal
                    (1, 0),   # Vertical
                    (1, 1),   # Diagonal (top-left to bottom-right)
                    (1, -1)   # Anti-diagonal (top-right to bottom-left)
                ]

                for delta_row, delta_col in directions:
                    if check_direction(row, col, delta_row, delta_col):
                        return get_cell(row, col)

    return 0
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import random
from kaggle_environments import evaluate, make, utils
```

```python

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
```

class PolicyNetwork(nn.Module):
    """Policy network that outputs action probabilities"""
    def __init__(self, input_size=42, hidden_size=256, output_size=7):
        super(PolicyNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, hidden_size)
        self.fc3 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)  # Return raw logits

```python
class REINFORCEAgent:
    """REINFORCE agent for ConnectX"""
    def __init__(self, lr=0.001, gamma=0.99):
        self.policy_net = PolicyNetwork()
        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=lr)
        self.gamma = gamma

        # Episode memory
        self.states = []
        self.actions = []
        self.log_probs = []
        self.rewards = []

    def get_valid_actions(self, board, columns=7):
        """Get list of valid column indices"""
        return [col for col in range(columns) if board[col] == 0]

    def select_action(self, board, valid_actions, training=True):
        """Select action using policy network"""
        # Convert board to tensor
        state = torch.FloatTensor(board).unsqueeze(0)

        # Get logits from policy network
        logits = self.policy_net(state)

        # Mask invalid actions by setting their logits to -inf
        masked_logits = logits.clone()
        for i in range(7):
            if i not in valid_actions:
                masked_logits[0][i] = float('-inf')

        # Convert to probabilities
        probs = F.softmax(masked_logits, dim=1)

        if training:
            # Sample action from distribution during training
            dist = torch.distributions.Categorical(probs)
            action = dist.sample()
            log_prob = dist.log_prob(action)
            return action.item(), log_prob
        else:
            # Take best action during evaluation
            return torch.argmax(probs).item(), None

    def store_transition(self, state, action, log_prob, reward):
        """Store transition in episode memory"""
        self.states.append(state)
        self.actions.append(action)
        self.log_probs.append(log_prob)
        self.rewards.append(reward)

    def calculate_returns(self):
        """Calculate discounted returns for each timestep"""
        returns = []
        G = 0

        # Work backwards from end of episode
        for reward in reversed(self.rewards):
            G = reward + self.gamma * G
            returns.insert(0, G)

        return torch.FloatTensor(returns)

    def update_policy(self):
        """Update policy using REINFORCE algorithm"""
        if len(self.rewards) == 0:
            return

        # Calculate returns
        returns = self.calculate_returns()

        # Normalize returns (optional but often helps)
        returns = (returns - returns.mean()) / (returns.std() + 1e-8)

        # Calculate policy loss
        policy_loss = 0
        for log_prob, G in zip(self.log_probs, returns):
            policy_loss += -log_prob * G

        # Update policy network
        self.optimizer.zero_grad()
        policy_loss.backward()
        self.optimizer.step()

    def clear_episode_memory(self):
        """Clear episode memory for next episode"""
        self.states.clear()
        self.actions.clear()
        self.log_probs.clear()
        self.rewards.clear()
```

```python
def play_episode(agent, env_config, as_player_1=True):
    """Play a single episode of ConnectX with self-play"""
    rows = env_config['rows']
    columns = env_config['columns']

    # Initialize board
    board = [0] * (rows * columns)
    current_player = 1
    episode_moves = []  # Store moves for both players

    while True:
        valid_actions = agent.get_valid_actions(board, columns)

        if not valid_actions:
            # Board is full - draw
            reward = 0
            break

        # Current player makes move
        action, log_prob = agent.select_action(board, valid_actions, training=True)

        # Store move if it's our training player
        if (as_player_1 and current_player == 1) or (not as_player_1 and current_player == 2):
            episode_moves.append({
                'board': board.copy(),
                'action': action,
                'log_prob': log_prob,
                'player': current_player
            })

        # Make the move
        board = make_move(board, action, current_player, rows, columns)

        # Check for win
        if check_win(board, current_player, rows, columns, env_config['inarow']):
            if (as_player_1 and current_player == 1) or (not as_player_1 and current_player == 2):
                reward = 1  # We won
            else:
                reward = -1  # We lost
            break

        # Switch players
        current_player = 3 - current_player

    # Store transitions for our player only
    for i, move in enumerate(episode_moves):
        agent.store_transition(
            move['board'],
            move['action'],
            move['log_prob'],
            reward if i == len(episode_moves) - 1 else 0  # Only final reward
        )

    return reward
```

```python
def make_move(board, action, player, rows, columns):
    """Make a move on the board"""
    board = board.copy()
    # Find the lowest empty row in the chosen column
    for row in range(rows):
        index = (rows - 1 - row) * columns + action
        if board[index] == 0:
            board[index] = player
            break
    return board
```

```python
def check_win(board, player, rows, columns, inarow):
    """Check if player has won"""
    # Check horizontal, vertical, and diagonal wins
    for row in range(rows):
        for col in range(columns):
            if board[row * columns + col] == player:
                # Check right
                if col <= columns - inarow:
                    if all(board[row * columns + col + i] == player for i in range(inarow)):
                        return True

                # Check down
                if row <= rows - inarow:
                    if all(board[(row + i) * columns + col] == player for i in range(inarow)):
                        return True

                # Check diagonal down-right
                if row <= rows - inarow and col <= columns - inarow:
                    if all(board[(row + i) * columns + col + i] == player for i in range(inarow)):
                        return True

                # Check diagonal down-left
                if row <= rows - inarow and col >= inarow - 1:
                    if all(board[(row + i) * columns + col - i] == player for i in range(inarow)):
                        return True

    return False
```

```python
def train_agent(episodes=10000):
    """Train the REINFORCE agent"""
    # Initialize Kaggle environment to get configuration
    env = make("connectx", debug=False)
    env_config = env.configuration

    agent = REINFORCEAgent(lr=0.001, gamma=0.99)

    win_rate_window = []

    for episode in range(episodes):
        # Alternate between playing as player 1 and player 2
        as_player_1 = episode % 2 == 0

        # Play episode
        reward = play_episode(agent, env_config, as_player_1)

        # Update policy after each episode
        agent.update_policy()
        agent.clear_episode_memory()

        # Track performance
        win_rate_window.append(1 if reward == 1 else 0)
        if len(win_rate_window) > 100:
            win_rate_window.pop(0)

        # Print progress
        if episode % 1000 == 0:
            avg_win_rate = sum(win_rate_window) / len(win_rate_window) if win_rate_window else 0
            print(f"Episode {episode}, Win Rate (last 100): {avg_win_rate:.3f}")

    return agent
```

```python
def create_kaggle_agent(agent):
    """Create agent function for Kaggle submission"""
    def kaggle_agent(obs, config):
        board = obs.board
        valid_actions = [col for col in range(config.columns) if board[col] == 0]

        if not valid_actions:
            return random.choice(range(config.columns))

        action, _ = agent.select_action(board, valid_actions, training=False)
        return action

    return kaggle_agent
```

```python
# Training and Evaluation
if __name__ == "__main__":
    print("Training REINFORCE agent...")
    trained_agent = train_agent(episodes=10000)

    # Save the trained model
    torch.save(trained_agent.policy_net.state_dict(), 'reinforce_connectx_model.pth')
    print("Model saved as 'reinforce_connectx_model.pth'")

    # Test against random agent
    print("\nTesting against random agent...")
    kaggle_agent = create_kaggle_agent(trained_agent)

    env = make("connectx", debug=False)

    # Test as player 1
    results_p1 = evaluate("connectx", [kaggle_agent, "random"], num_episodes=100)
    wins_p1 = sum([1 for r in results_p1 if r[0] == 1])

    # Test as player 2
    results_p2 = evaluate("connectx", ["random", kaggle_agent], num_episodes=100)
    wins_p2 = sum([1 for r in results_p2 if r[1] == 1])

    print(f"Win rate as Player 1: {wins_p1}%")
    print(f"Win rate as Player 2: {wins_p2}%")
    print(f"Overall win rate: {(wins_p1 + wins_p2)/2}%")
```
